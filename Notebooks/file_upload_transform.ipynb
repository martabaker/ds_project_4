```python
# Dependencies
import pandas as pd
```

```python
# Import Files
test_df = pd.read_csv("../Resources/test.csv", encoding="UTF-8")
train_df = pd.read_csv("../Resources/train.csv", encoding="UTF-8")

display(test_df.head())
train_df.head()
```

### Data Cleaning

```python
# Fix column names
test_df.columns = [x.lower().replace(" ", "_") for x in test_df.columns]
train_df.columns = [x.lower().replace(" ", "_") for x in train_df.columns]

test_df.rename(columns={"class": "travel_class"}, inplace=True)
train_df.rename(columns={"class": "travel_class"}, inplace=True)
```

```python
# Determine datatypes and check for nulls
display(test_df.info())
train_df.info()
```

Note: There appear to be some nulls in `arrival_delay_in_minutes` for both tables

```python
# Check percentage of nulls for each dataset
print(f"test nulls: {(25976-25893)/25976} \ntrain nulls: {(103904-103594)/103904}")
```

```python
# Given how few nulls there are, going to just drop those rows
test_df = test_df.dropna(how="any").reset_index(drop=True)
train_df = train_df.dropna(how="any").reset_index(drop=True)

display(test_df.head())
train_df.head()
```

Drop the `unnamed:_0` column because it's another index column and as such is redundant

```python
test_df.drop(columns=test_df.columns[0], axis=1, inplace=True)
train_df.drop(columns=train_df.columns[0], axis=1, inplace=True)
```

### Combine both tables to create a complete table

```python
combined_df = pd.concat([test_df, train_df])
display(combined_df.shape)
display(combined_df.head())
combined_df.info()
```

### Save Off Tables

```python
test_df.to_csv("../Resources/test_cleaned.csv", index=False)
train_df.to_csv("../Resources/train_cleaned.csv", index=False)
combined_df.to_csv("../Resources/combined_cleaned.csv", index=False)
```
